2. **Experiment:** Rolling a fair die, **Event:** A number greater than 3 occurs.

```python
a = np.random.randint(1,7,size=10000)
k = 0
arr = []
for i in range(len(a)):
    if a[i]>3:
        k+=1
        val = k/(i+1)
        arr.append(val)
        
plt.plot(arr)     
```

3. **Experiment:** Rolling two fair dice, **Event:** Sum of two dice equal 5.

```python
a1 = np.random.randint(1,7,size=10000)
a2 = np.random.randint(1,7,size=10000)
q = 0
arr_get = []
for i in range(len(a1)):
    sum_dies = a1[i] + a2[i]
    if sum_dies == 5:
        q+=1
        val = q/(i+1)
        arr_get.append(val)

print('final Result is ... : ',arr_get[-1])
plt.plot(arr_get)
```

# Problem 2

Simulate the experiment of tossing $1000$ times a coin with probability of **head (coded as $1$) equal to $0.4$** and probability of **tail (coded as $0$) equal to $0.6$**. How can you check if the simulation of the experiment of been running as expected.

**Hint**: Use [matplotlib.pyplot.hist¶
](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.hist.html)

```python
import random
def biased_coin(p):
    b_coin = random.random()
    return b_coin <= p
cHead = 0
cTail = 0

for i in range(1000):
    if biased_coin(0.4):
        cHead+=1
    if biased_coin(0.6):
        cTail+=1

list1 = [cTail,cHead]
list2 = [0,1]
plt.bar(list2,list1 , align = 'edge', width=[0.12,0.12])

plt.show()
```

```python
# other solution

a = np.random.rand(1000)
c_head = 0
c_tail = 0

for i in range(len(a)):
    if a[i] < 0.4:
        c_head+=1
    if a[i] < 0.6:
        c_tail+=1
        
y_res = [c_tail,c_head]
x_axis = [0,1]
plt.bar(x_axis,y_res , align = 'edge', width = [0.12,0.12])
plt.show()
```
